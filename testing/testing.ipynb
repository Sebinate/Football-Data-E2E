```python
import pandas as pd
import requests
from dotenv import load_dotenv
import os
```

```python
host = os.getenv('FOOTBALL_API_HOST')
api = os.getenv('FOOTBALL_API_KEY')

headers={'x-apisports-key': api}
```

### League

```python
target1 = 'leagues'

response = requests.get(url = f'{host}/{target1}', headers=headers)
data = response.json().get('response')

dataframe = pd.json_normalize(data)
dataframe.head()
```

```python
dataframe
```

## Countries

## Teams

```python
dataframe
```

```python
prem_index = dataframe.loc[(dataframe['league.name'] == 'Premier League') & (dataframe['country.name'] == 'England') , 'league.id'].to_string(index = False)
```

```python
target3 = 'teams'

response3 = requests.get(url=f'{host}/{target3}', headers=headers, params={'league': prem_index, 'season': '2022'})
data3 = response3.json().get('response')

data3
```

```python
dataframe3 = pd.json_normalize(data3)
dataframe3.head()
```
